```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/Caiooliveirabarreto/Aulas_PA/refs/heads/main/An%C3%A1lise%20de%20dados/AB_NYC_2019.csv')
```

```python
data.head(3)
```

```python
#id = ID de listagem
#name = nome da listagem
#host_id = ID do dono
#host_name = Nome do dono
#neighbourhood_group = Área geral
#neighbourhood = Área especializada
#latitude = coordernadas de latitude
#longitude = coordenadas de longitude
#room_type = listagem do tipo de lugar
#price = Preço em dólares
#minimum_nights = mínimo de noites alugaveís
#number_of_reviews = número de avaliações
#last_review = última avaliação
#reviews_per_month = avaliações por mês
#calculated_host_listings_count = número de listagens feitas pelo dono
#availability_365 = disponibilidade por dias
```

```python
print(data['reviews_per_month'].mean()) # Média
```

```python
data['reviews_per_month'] = data['reviews_per_month'].fillna(data['reviews_per_month'].mean()) # Média preenchida nos dados
```

```python
data.isnull().sum()
```

```python
data.head()
```

```python
# Limpeza de dados
data.drop('name', axis=1, inplace=True)
data.drop('host_id', axis=1, inplace=True)
data.drop('host_name', axis=1, inplace=True)
data.drop('id', axis=1, inplace=True)
data.drop('last_review', axis=1, inplace=True)
data = data.dropna()
```

```python
sns.color_palette(palette='crest')
sns.set_palette("crest")
sns.barplot(data, y='number_of_reviews', x='neighbourhood_group', hue='neighbourhood_group')
```

É possível se observar que uma quantidade a mais de ímoveis em Staten Island, que é acompanhado por Queens e pelo Bronx

```python
sns.barplot(data, y='price', x='neighbourhood_group', hue='neighbourhood_group')
```

Apesar de Manhattan ter menos imóveis, é possível perceber que seus imóveis tem um preço um maior, indicando imóveis mais luxuosos, ou imóveis com uma relação custo-benefício ruím

\
O Bronx tem o menor preço da região, indicando menos imóveis de luxo.

```python
sns.barplot(data, y='number_of_reviews', x='room_type', hue='room_type')
```

É possível se perceber que a maioria das pessoa preferem uma quarto privado, e até preferem uma casa inteira a um quarto compartilhado
