# Coletando os valores para as variaveis de entrada

```python
distancia = float(input("digite a distância percorrida: "))
qtde_horas = int(input("digite a quantidade de horas gastas: "))
qtde_min = int(input("digite a quantidade de minutos gastos: "))
```

```python
# Processamento
tempo_total = qtde_horas + (qtde_min/60)
velocidade = distancia/tempo_total
```

```python
velocidade = round(velocidade, 2)
print("sua velocidade média é: ", velocidade,"km/h")
```
